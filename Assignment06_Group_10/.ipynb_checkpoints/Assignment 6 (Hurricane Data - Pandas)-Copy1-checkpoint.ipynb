#### The month_to_quarter function maps a month name to a corresponding quarter. It takes a month as input and returns the quarter as a string. 

```python
# Define function to map months to quarters
def m_to_q(m):
    if m in ['January', 'February', 'March']:
        return 'Spring'
    elif m in ['April', 'May', 'June']:
        return 'Summer'
    elif m in ['July', 'August', 'September']:
        return 'Fall'
    else:
        return 'Winter'
```

```python
hurri_df['quarter'] = hurri_df['month'].apply(m_to_q)
```

#### Groups the DataFrame df by the 'year' and 'quarter' columns, and then calculates the mean and standard deviation of the 'hurricanes' column for each group.

```python
# Group by year and quarter, then calculate mean and standard deviation
yearly_quarterly_stat = hurri_df.groupby(['year', 'quarter'])['hurricanes'].agg(['mean', 'std']).rename(columns={'mean': 'Average', 'std': 'Standard Deviation'})
```
